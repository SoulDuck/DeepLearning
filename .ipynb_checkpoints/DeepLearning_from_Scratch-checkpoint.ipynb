```python
def AND(x1 , x2):
    w1 , w2 , theta = 0.5 , 0.5 , 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0;
    elif tmp>theta:
        return 1
```

```python
print AND(0,0)
print AND(1,0)
print AND(0,1)
print AND(1,1)
```

###### 그럼 W1 ,W2 은 어떤 기능을 할까?


Weight : 입력신호가 결과에 주는 영향력을 조절하는 매개변수 

Bias : 뉴런이 얼마나 쉽게 활성화 하느냐를 조정하는 매개변수

e.g) if b = -1 이면 W*X > -1 이상이여야 Neuren이 활성화 된다.

###### 그럼 NAND gate 와 OR Grpah 는 어떨까?

```python
def NAND(x1 ,x2):
    x = np.array([x1, x2])
    w = np.array([-0.5 , -0.5])
    b = 0.7
    
    tmp = np.sum(w*x)+b
    if tmp <=0:
        return 0
    else:
        return 1
    
```

```python
def OR(x1 , x2):
    x = np.array([x1 , x2])
    w = np.array([ 0.5 , 0.5])
    b = -0.2
    tmp = np.sum(x*w) +b
    if tmp<= 0:
        return 0
    else:
        return 1
```

```python
print NAND(0,0)
print NAND(0,1)
print NAND(1,0)
print NAND(1,1)
```

```python
print OR(0,0)
print OR(0,1)
print OR(1,0)
print OR(1,1)
```

```python
def XOR(x1,x2):
    s1=NAND(x1,x2)
    s2=OR(x1,x2)
    y=AND(s1,s2)
    return y
```

```python
print XOR(0,0)
print XOR(1,0)
print XOR(0,1)
print XOR(1,1)
```

###### XOR
NAND : x1*w1+x2*w2+b1 OR:x1*w3+x2*w4+b2

w1:-0.5 w2:-0.5 b1:0.7 w3:0.5 w4:0.5 b2:-0.2

AND: (x1*w1+x2*w2+b1)*w5+(x1*w3+x2*w4+b2)*w6+b3

w5: 0.5 w6 : 0.5  b3 :0.7

```python
plt.plot([0,1] , [0,1] , 'ro')
plt.plot([0,1] ,[1,0] , 'bs')
plt.xlim(-1,2)
plt.ylim(-1,2)
x=range(-10,20)
x_np=np.asarray(x)*0.1
x=list(x_np)
y=x_np*-0.5+0.7
y2 = x_np*0.5-0.2
y3 = x_np*0.5+0.7
plt.plot(x , y)
plt.plot(x , y2)
plt.plot(x, y3)
plt.title('XOR')
```

###### 입력값에 의해 Weight , Bias 가 정해진다.

###### Purpose:   Deep Learning을 이용해   Weights , Biases을 학습시킨다


###### How?  WX+B 의 예상값 과 실제값의 차이를 줄이는 방향으로 학습을 시킨다.


######  WX+B 의 예상값 과 실제값의 차이를  어떻게 만드는지를 알아야 합니다.

### 1.Step Function

```python
def step(x):
    y=x>0
    y=y.astype(np.int)
    return y
```

```python
x=np.arange(-2.0 , 2.0 ,0.001)
y=step(x)
plt.ylim(-2,2)
plt.plot(x,y)
```

###### Sigmoid function

```python
def Sigmoid(x):
    return 1 / (np.exp(-x)+1)
```

```python
def exponential(x):
    return np.exp(x)
```

```python
x=np.arange(-2 , 2 , 0.0001)
y=exponential(x)
plt.plot(x,y)
plt.title('exponential')
```

```python
def minus_exponential(x):
    return np.exp(x)*-1
```

```python
x=np.arange(-2 , 2 , 0.0001)
y=minus_exponential(x)
plt.plot(x,y)
plt.title('exponential')
```

```python
def minus_exponential_recipocal(x):
    return 1/np.exp(x)*-1
```

```python
x=np.arange(-2 , 2 , 0.0001)
y=minus_exponential_recipocal(x)
plt.plot(x,y)
plt.title('minus exponential recipocal')
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x=np.arange(-5 , 5 , 0.0001)
y=relu(x)
plt.plot(x,y)
plt.ylim(-2,2)
plt.xlim(-2,2)
plt.title('Relu')
```
